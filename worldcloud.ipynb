aplicar a un texto:

```python
text = "Natural Language Processing with NLTK and Python is fun and powerful!"
```

```python
wc = WordCloud()
```

```python
type(wc)
```

```python
wc = WordCloud(
    background_color="white",
    stopwords=STOPWORDS,
    max_font_size=40
).generate(text)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.imshow(wc)
plt.show()
```

aplicar a un texto más largo:

librerías:

```python

from nltk.corpus import gutenberg
```

```python
text = gutenberg.raw("shakespeare-hamlet.txt")
```

```python
wc = WordCloud(
    background_color="white",
    max_words=100,
    max_font_size=40,
    stopwords=STOPWORDS,
    random_state=96,
    repeat=False,
    colormap="gist_rainbow",
    min_word_length=4
).generate(text)
```

```python
plt.figure(figsize=(10, 6))
plt.imshow(wc, interpolation="bilinear")
plt.axis("off")
plt.show()
```
