calcular tf

```python
def calcular_tf(term:str, doc:list):
    apariciones = doc.count(term)
    return apariciones / len(doc)
```

calcular idf

```python
def calcular_idf(term:str, corpus:list):
    tf = sum([1 for doc in corpus if term in doc])
    return log((1 + len(corpus)) / (1 + tf)) + 1
```

calcular tfidf

```python
def calcular_tfidf(term:str, doc:list, corpus:list):
    tf = calcular_tf(term, doc)
    idf = calcular_idf(term, corpus)
    return tf * idf
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer


tfidf_vectorizer = TfidfVectorizer(token_pattern=r"\w+", norm=None)
```

```python
tfidf_matrix = tfidf_vectorizer.fit_transform(Docs)
```

```python
print(f"Forma de la matriz: {tfidf_matrix.shape}")
```

```python
vocab = tfidf_vectorizer.get_feature_names_out()

print(f"Vocabulario: {vocab}")
print(f"Tamaño del vocabulario: {len(vocab)}")
```

tabla donde los documentos son filas y las columnas palabras del vocabulario y contienen los valors td idf de las palabras

```python
pd.DataFrame(tfidf_matrix.toarray(), columns=vocab, index=["d1", "d2", "d3"])
```

comparar el parecido entre textos

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similitud = cosine_similarity(tfidf_matrix)
```

```python
tmp = ["d1", "d2", "d3"]
pd.DataFrame(similitud.round(3), index=tmp, columns=tmp)
```
