```python
from sklearn.datasets import fetch_20newsgroups
```

cargar dataset con noticias de categorias de baseball y autos

```python
noticias_2gpos = fetch_20newsgroups(categories=['rec.sport.baseball', 'rec.autos'])
```

```python
type(noticias_2gpos)
```

```python
len(noticias_2gpos.data)
```

a que categoría pertenece cada noticia?

arreglo de etiquetas numéricas que muestran a qué categoría pertenece cada noticia

```python
noticias_2gpos.target
```

separar los datos de las etiquetas:

```python
X = noticias_2gpos.data
y = noticias_2gpos.target
```

```python
print(len(X), len(y))
```

```python
from pandas import DataFrame
```

asignamos las etiquetas y texto que habíamos separado

```python
df = DataFrame(
    {"Noticia": X,
     "Etiqueta": y}
)
```

```python
df.head()
```

convertir los textos de las noticias en números

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizador = TfidfVectorizer()
X_vect = vectorizador.fit_transform(df.Noticia)
X_vect
```
